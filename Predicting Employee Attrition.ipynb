# Label Encoding

```python
#target variable(Attrition)
data['Attrition'] = data['Attrition'].replace({'Yes':1,'No':0})
```

```python
# encode binary variables
data['OverTime'] = data['OverTime'].map({'Yes':1,'No':0})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
data['Over18'] = data['Over18'].map({'Y':1,'N':0})
```

```python
# encoding categorical columns which are ordinal, use labelEncoding
# applying label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
encoding_cols = ['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders ={}
for column in encoding_cols:
  label_encoders[column] = LabelEncoder()
  data[column] = label_encoders[column].fit_transform(data[column])
```

# Machine learning splitting the data into training and testing sample

```python
x = data.drop(['Attrition'],axis=1)
y = data['Attrition'].values
```

# Resampling

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state=42)
x, y = rus.fit_resample(x, y)
print(Counter(y))
```

```python
# SPlit the data into training and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Sanity check for the sampled data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Logistic Regression In machine Learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,roc_curve
```

```python
from math import log
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
Prediction = log_reg.predict(x_test)
cnf_matrix = confusion_matrix(y_test,Prediction)
print("Accuracy Score:",accuracy_score(y_test,Prediction))
```

```python
fig = plt.figure(figsize=(15,6))
ax1=fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues', fmt ='d')
bottom, top, = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
ax1.set_title('Confusion Matrix')
ax1.set_xlabel('Predicted')
ax1.set_ylabel('Actual')

ax2 = fig.add_subplot(1,2,2)
y_pred_prob = log_reg.predict_proba(x_test)[::,1]
fpr, tpr, _ = roc_curve(y_test, Prediction)
auc = roc_auc_score(y_test, Prediction)
ax2 = plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.legend(loc =4)
plt.show()
```
